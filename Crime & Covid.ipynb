```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\Bizhan\Desktop\hate_crime.csv"

df = pd.read_csv(file_path)
```

```python
print(df.head())
```

```python
df_2019 = df[df["data_year"] == 2019]

# Calculate the total number of incidents
total_incidents = df_2019["incident_id"].sum()

print("The total number of incidents for 2019 is:", total_incidents)
```

```python
df_2020 = df[df["data_year"] == 2020]

# Calculate the total number of incidents
total_incidents = df_2020["incident_id"].sum()

print("The total number of incidents for 2020 is:", total_incidents)
```

```python
df_2020 = df[df["data_year"] == 2021]

# Calculate the total number of incidents
total_incidents = df_2020["incident_id"].sum()

print("The total number of incidents for 2021 is:", total_incidents)
```

```python
df_1991_2022 = df[(df["data_year"] >= 1991) & (df["data_year"] <= 2022)]

# Group the data by year and calculate the total number of incidents
yearly_incidents = df_1991_2022.groupby("data_year")["incident_id"].sum()

# Create a histogram of the yearly incidents
print(yearly_incidents)
```

```python
# Create a histogram of the yearly incidents
plt.figure(figsize=(16, 8))
plt.bar(yearly_incidents.index, yearly_incidents.values)
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.title('Total Number of Hate Crime Incidents per Year (1991-2022)')
plt.xticks(yearly_incidents.index)
plt.show()
```

```python

import seaborn as sns

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")

plt.title("Correlation Matrix")
plt.show()
```

```python
df_2022 = df[df["data_year"] == 2022]
state_incidents = df_2022.groupby("state_name")["incident_id"].sum()

top_10_states = state_incidents.sort_values(ascending=False)[:10]

# Creating a pie chart
plt.figure(figsize=(16, 8))
plt.pie(top_10_states.values, labels=top_10_states.index)
plt.title('Top 10 States with Most Number of Hate Crime Incidents in 2022')
plt.show()
```
